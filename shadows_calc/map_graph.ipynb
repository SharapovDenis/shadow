```python
import numpy as np
import matplotlib.pyplot as plt

data = np.recfromcsv('output_data.csv')
data
```

```python
lat, lon = [], []

i = 0
while i < len(data) and i < 60000:
    house = data[i][1]
    
    latt, lonn = [], []
    while i < len(data) and house == data[i][1]:
        latt.append(data[i][3])
        lonn.append(data[i][4])
        i += 1

    lat.append(lonn)
    lon.append(latt)

print(i, ' / ', data[-1][0])
```

```python
fig, ax = plt.subplots(figsize=(12.8, 7.2), dpi=110)

for x, y in zip(lat, lon):
    ax.plot(x, y, lw=0.5, c='tab:red')

print(len(lat))

plt.show()
# plt.savefig('map.png', dpi=200)
```
