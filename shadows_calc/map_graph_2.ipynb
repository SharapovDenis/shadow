```python
import numpy as np
import matplotlib.pyplot as plt

data = np.recfromcsv('output_data.csv')
data_new = np.recfromcsv('output_data_2.csv')

print(data)
print(data_new)
```

```python
i_max = 60000
```

```python
lat, lon = [], []

i = 0
while i < len(data) and i < i_max:
    house = data[i][1]

    latt, lonn = [], []
    while i < len(data) and house == data[i][1]:
        latt.append(data[i][3])
        lonn.append(data[i][4])
        i += 1

    lat.append(lonn)
    lon.append(latt)

print(i, ' / ', data[-1][0])
```

```python
lat_new, lon_new = [], []

i = 0
while i < len(data_new) and i < i_max:
    house = data_new[i][1]

    latt_new, lonn_new = [], []
    while i < len(data_new) and house == data_new[i][1]:
        latt_new.append(data_new[i][3])
        lonn_new.append(data_new[i][4])
        i += 1

    lat_new.append(lonn_new)
    lon_new.append(latt_new)

print(i, ' / ', data_new[-1][0])
```

```python
fig, ax = plt.subplots(figsize=(12.8, 7.2), dpi=110)

for x, y, x_new, y_new in zip(lat, lon, lat_new, lon_new):

    xx, yy = [], []
    for i in range(len(x)):
        xx.append(x[i])
        xx.append(x_new[i])
        yy.append(y[i])
        yy.append(y_new[i])

    ax.plot(xx, yy, lw=1, c='black')
    ax.plot(x, y, lw=1, c='black')
    ax.plot(x, y, lw=0.5, c='white')


print(len(lat))

plt.show()
# plt.savefig('map2.png', dpi=200)
```
